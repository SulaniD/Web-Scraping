##  Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Fetch the webpage content
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"  # URL containing Tesla's revenue data
response = requests.get(url)

# Step 2: Parse the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Find the table containing revenue data
# We will use a generic approach to find the first table
table = soup.find('table')

# Step 4: Extract table rows
rows = table.select('tr')  # Selects all <tr> elements in the table


# Step 5: Process the rows into a list of dictionaries
data = []
for row in rows[1:]:  # Skip the header row
    columns = row.find_all('td')
    if len(columns) > 1:
        year = columns[0].text.strip()
        revenue = columns[1].text.strip()
        data.append({'Year': year, 'Revenue': revenue})

# Step 6: Create a DataFrame from the list of dictionaries
tesla_revenue = pd.DataFrame(data)

# Step 7: Display the last five rows using the tail() function
print(tesla_revenue.tail())



```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Tesla stock data from January 2020 to January 2025
tsla_data = yf.download('TSLA', start='2020-01-01', end='2025-01-01')

# Define the make_graph function
def make_graph(stock_data):
    # Plot the closing price of Tesla stock
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Close'], label='Tesla Closing Price', color='blue')
    plt.title('Tesla Stock Price (TSLA)', fontsize=14)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

# Call the make_graph function to display the graph
make_graph(tsla_data)
```
